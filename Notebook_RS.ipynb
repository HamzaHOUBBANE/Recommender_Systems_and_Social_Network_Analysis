```python
def read_chunks(file, cols, chunk_size=500000):
    df = pd.read_json(
        f'C:\yelp_dataset\yelp_academic_dataset_{file}.json', chunksize = chunk_size, lines=True)
    
    chunk_list = [chunk[cols] for chunk in df ]

    return pd.concat(chunk_list, ignore_index = True, join = 'outer', axis = 0)
```

```python
df_user = read_chunks('user', ['user_id','name','friends','fans'])
```
