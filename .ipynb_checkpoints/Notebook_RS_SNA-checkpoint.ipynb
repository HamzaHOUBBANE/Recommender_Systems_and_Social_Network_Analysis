```python
def read_chunks(file, cols, chunk_size=500000):
    df = pd.read_json(
        f'C:\yelp_dataset\yelp_academic_dataset_{file}.json', chunksize = chunk_size, lines=True)
    
    chunk_list = [chunk[cols] for chunk in df ]

    return pd.concat(chunk_list, ignore_index = True, join = 'outer', axis = 0)
```

```python
# Sample 20000 rows from the DataFrame
dfu_franklin_sample = dfu_franklin.sample(n=20, random_state=1)

# Create a graph
G = nx.Graph()

# Add nodes and edges based on your criteria
for index, row in dfu_franklin_sample.iterrows():
    user_id = row['user_id']
    G.add_node(user_id)
    if isinstance(row['friends'], str):
        friends = row['friends'].split(', ')
        for friend in friends:
            G.add_edge(user_id, friend)

# Draw the graph
plt.figure(figsize=(10,10))
nx.draw_networkx(G, with_labels=False, node_size=1)
plt.show()
```

```python
from networkx.algorithms import community
# Sample 20000 rows from the DataFrame
dfu_franklin_sample = dfu_franklin.sample(n=20, random_state=1)

# Create a graph
G = nx.Graph()

# Add nodes and edges based on your criteria
for index, row in dfu_franklin_sample.iterrows():
    user_id = row['user_id']
    G.add_node(user_id)
    if isinstance(row['friends'], str):
        friends = row['friends'].split(', ')
        for friend in friends:
            G.add_edge(user_id, friend)

# Find communities
communities_generator = community.girvan_newman(G)
top_level_communities = next(communities_generator)
communities = sorted(map(sorted, top_level_communities))

# Create a color map for communities
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
color_map = {}

for i, community in enumerate(communities):
    for node in community:
        color_map[node] = colors[i % len(colors)]

# Draw the graph with different colors for each community
plt.figure(figsize=(10,10))
nx.draw_networkx(G, with_labels=False, node_size=1, node_color=[color_map[node] for node in G.nodes()])
plt.show()
```
